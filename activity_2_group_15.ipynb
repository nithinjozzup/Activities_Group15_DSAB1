# 1.Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.

```python
diabm='diabetes_model.csv'
df1=pd.read_csv(diabm)
df2=df1.loc[45:60,['Glucose','Age']]
df2
```

# 2. How many patients in the data are having the blood pressure level above 120?

```python
df2=df1.loc[:,['BloodPressure']]>120
df3=df1.loc[df1['BloodPressure']>120]
patient=df2.sum()
df3
```

```python
print("\n Number of patients with BP above 120=",patient['BloodPressure'])
```

# 3. Find the minimum and maximum value of insulin level in the data.

```python
maxin=df1.loc[:,['Insulin']].max()
minin=df1.loc[:,['Insulin']].min()
print('The minimum and maximum value of insulin level in the data are:\n Min:',minin['Insulin'],'\n Max:',maxin['Insulin'])
```

# 4. Create a histogram of "Age".

```python
plt.figure(figsize=(10,10))
plt.hist(df1['Age'],color='green',rwidth=0.9)
plt.title('Frequency of age')
plt.xlabel('AGE')
plt.ylabel('Frequency')
```

# 5. Create a scatter plot of "BMI" vs "Glucose".

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,10))
sns.scatterplot(data=df1,x ='BMI', y="Glucose")
plt.title('BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
```
