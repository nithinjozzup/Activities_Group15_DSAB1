# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the dataset

```python
tel='Tel.xlsx'
df=pd.read_excel(tel)
df
```

# 1) Compare churn count with respect to gender.

```python
plt.figure(figsize=(20,10))

sns.countplot(x='gender',data=df,hue="Churn")
plt.title('Gender vs Churn count',size=20)
```

# Insights
    1. Similar churn count for male and female
    2. People who opted for churn is 1/3 of the people who opted out.

# 2) Findout how many female senior citizen there in the dataset

```python
df2= df[(df['gender']=='Female') & (df['SeniorCitizen']==1) ]
df2    
        
```

```python
count=0
for i in df2['gender']:
    count+=1
print('female senior citizen in the dataset:',count)
```

# 3) Compare 'tenure' with 'Total Charges'

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors = 'coerce')
plt.figure(figsize=(20,10))
sns.scatterplot(x='tenure',y='TotalCharges',data=df)
plt.title('Tenure vs TotalCharges',size=20)
```

# Insights
    1. It is noted that there is a proportional increase for total charges as the tenure is increased.

# 4) Findout which contract preffered by the senior citizen.

```python
plt.figure(figsize=(20,10))

sns.countplot(x='SeniorCitizen',data=df,hue="Contract")
plt.title('SeniorCitizen prefered contract',size=20)
```

# Insight
    1. Senior citizen prefers Month-to-month contract.

# 5) Comment your finds on Payment Method?

```python
plt.figure(figsize=(20,10))

sns.countplot(x='SeniorCitizen',data=df,hue="PaymentMethod")
plt.title('Payment method based on Citizen category ',size=20)

plt.figure(figsize=(20,10))

sns.countplot(x='gender',data=df,hue="PaymentMethod")
plt.title('Payment method based on Gender',size=20)
```

# Insight
    1. Both all citizens generally use electronic check.
    2. Senior citizen least prefers mailed check although others have second preference for mailed check.
    3. Both male and female prefer Electronic check the most.
    4. While credit card is the least prefered in female ,Bank transfer is the least prefered in Male.
